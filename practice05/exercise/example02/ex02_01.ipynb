```python
y = df["class"]
X = df.drop("class", 1)
# Standardize the data
X_standard = (X - X.mean()) / X.std(ddof=0)
# Calculating the correlation matrix of the data
X_corr = (1 / 150) * X.T.dot(X)
# Plotting the correlation matrix
# plt.figure(figsize=(10,10))
# sns.heatmap(X_corr, vmax=1, square=True,annot=True)
# plt.title('Correlation matrix')
```

```python
pca_02 = PCA(n_components=2)  
X_pca_02 = pca_02.fit_transform(X_standard)  
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_pca_02[:, 0], X_pca_02[:, 1], c=y, cmap='viridis', alpha=0.5)
plt.title('Biểu đồ quan sát sau khi giảm số chiều về 2')
plt.colorbar(label='Class')
plt.show()
```
