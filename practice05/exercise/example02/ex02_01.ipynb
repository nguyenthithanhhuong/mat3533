Khai báo các thư viện cần thiết

```python
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

Đọc dữ liệu

```python
warnings.filterwarnings('ignore')
df= pd.read_csv("D:/Code/python/mat3533/practice05/data/pd_speech_features.csv")
df.head()
df['class'].value_counts()
```

Mô tả dữ liệu

```python
df.describe()
```

Chuẩn hóa dữ liệu

```python
y = df["class"]
X = df.drop("class", 1)

X_standard = (X - X.mean()) / X.std(ddof=0)
```

Đưa dữ liệu về còn 2 chiều

```python
pca_02 = PCA(n_components=2)  
X_pca_02 = pca_02.fit_transform(X_standard)  
```

Hiển thị lên màn hình để xem quan hệ giữa các lớp dữ liệu

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_pca_02[:, 0], X_pca_02[:, 1], c=y, cmap='viridis', alpha=0.5)
plt.title('Biểu đồ quan sát sau khi giảm số chiều về 2')
plt.colorbar(label='Class')
plt.show()
```
