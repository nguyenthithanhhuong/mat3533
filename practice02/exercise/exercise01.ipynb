```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:/Code/python/mat3533/practice02/data/SAT_GPA.csv")
data.describe()
```

```python
# Set to trainning data 
y = data['GPA']
x = data['SAT']
```

```python
# Remind that we need to put component x_0 = 1 to x
plt.scatter(x,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = np.array([x]).T
y = np.array([y]).T
x_training = x[:60]
y_training = y[:60]
x_validation = x[60:]
y_validation = y[60:]
```

```python
# Tisnh hệ số của mô hình tuyến tính
one = np.ones((x_training.shape[0], 1))
xbar = np.concatenate((one, x_training), axis = 1)
A = np.dot(xbar.T, xbar)
b = np.dot(xbar.T, y_training)
w = np.dot(np.linalg.pinv(A), b)
print("w = ", w)
```

```python
w_0 = w[0][0]
w_1 = w[1][0]
```

```python
plt.scatter(x_training, y_training)
yhat = w_1*x_training + w_0
fig = plt.plot(x_training, yhat, lw=4, c='orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
y_predict = w_1*x_validation + w_0
print(y_predict)
```

```python
sum_square_residual = sum((y_validation - y_predict)**2)
print(sum_square_residual)
```
