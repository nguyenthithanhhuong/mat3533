```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("D:/Code/python/mat3533/practice02/data/real_estate.csv")
data.describe()
```

```python
x_data = data.drop(['No', 'Y house price of unit area'], axis = 1)
y_data = [data['Y house price of unit area']]
x = np.array(x_data)
y = np.array(y_data).T
```

```python
x_train = x[:350]
y_train = y[:350]
x_test = x[350:]
y_test = y[350:]
```

```python
one = np.ones((x_train.shape[0], 1))
Xbar = np.concatenate((one, x_train), axis = 1)
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y_train)
w = np.dot(np.linalg.pinv(A), b)
print("w = ", w)
```

```python
one = np.ones((y_test.shape[0], 1))
Xbar = np.concatenate((one, x_test), axis = 1)
y_predict = np.dot(Xbar, w)
print(y_predict)
```

```python
sum_squar_residual = sum((y_test - y_predict)**2)
print(sum_squar_residual)
```
