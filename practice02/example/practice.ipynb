```python
from __future__ import print_function
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score
import numpy as np
import pandas as pd
```

```python
def read_data(path):
    with open(path, "r") as f:
        lines = f.readlines()
        lines = [x.strip() for x in lines if not x.startswith("I")]
        data = [x.split(" ") for x in lines]
        data = np.array(data).astype(float)
        data = pd.DataFrame(data)
    return data
```

```python
data = read_data("D:/Code/python/mat3533/practice02/data/vidu3_lin_reg.txt")
print(data)
```

```python
x = data.iloc[:, 1:6]
y = data[6]
x = np.array(x)
y = np.array([y]).T
```

```python
print(x)
```

```python
print(y)
```

```python
x_train = x[:80]
y_train = y[:80]
x_test = x[80:]
y_test = y[80:]
```

```python
one = np.array([np.ones(len(x_train))]).T
xbar = np.concatenate((one, x_train), axis=1)
A = np.dot(xbar.T, xbar)
b = np.dot(xbar.T, y_train)
w = np.dot(np.linalg.pinv(A), b)
print(w)
```

```python
one_test = np.array([np.ones(len(x_test))]).T
xbar_test = np.concatenate((one_test, x_test), axis=1)
y_predict = np.dot(xbar_test, w)
print(y_predict)
```

```python
mean = sum(y_test - y_predict) / len(y_test)
print("Mean: ", mean)
variance = sum((y_test - mean)**2)
print("Variance: ", variance)   
```
