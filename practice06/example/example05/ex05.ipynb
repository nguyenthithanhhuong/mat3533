```python
import numpy as np  
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
data = pd.read_csv("D:/Code/python/mat3533/practice06/data/Analysis-Portfolio-Task-Data.csv")
data
```

```python
data['Gender'] = data['Gender'].map({'Male' : 0,'Female' : 1})
data['Store Type'] = data['Store Type'].map({'Convenient Stores' : 1,'Superstore' : 2, 'Online': 3})
data.drop('CustomerID', axis=1, inplace=True)
data.dropna(inplace=True)
data
```

```python
X = data.drop('Shopping Basket', axis=1)
y = data['Shopping Basket']
y = np.where(data['Shopping Basket'] > 50, 1, 0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
```

```python
X_train = X_standard[:60]
X_test = X_standard[60:]
y_train = y[:60]
y_test = y[60:]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score

model_log_reg = LogisticRegression()
model_log_reg.fit(X_train, y_train)

y_pred_log_reg = model_log_reg.predict(X_test)
print(f"Prediction:\n {y_pred_log_reg}")

accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
precision_log_reg = precision_score(y_test, y_pred_log_reg)
recall_log_reg = recall_score(y_test, y_pred_log_reg)

print("Accuracy: ", accuracy_log_reg)
print("Precision: ", precision_log_reg)
print("Recall: ", recall_log_reg)


coefficients = model_log_reg.coef_

# In ra các hệ số
print("Coeficients:")
for coef in coefficients:
    print(f"{coef} ")
```
