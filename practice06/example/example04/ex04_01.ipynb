Khai báo các thư viện cần thiêt

```python
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Đọc và xử lý dữ liệu trống

```python
data = pd.read_csv('D:/Code/python/mat3533/practice06/data/framingham.csv')
data = data.dropna()
data.describe()
```

```python
X = data.drop(['TenYearCHD'], axis=1)
y = data['TenYearCHD']
```

Giảm số chiều của dữ liệu về 2 chiều

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

Hiển thị trực quan dữ liệu sau khi giảm số chiều

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='bwr', alpha=0.5)
plt.title('Biểu đồ hiển thị dữ liệu sau khi giảm về 2 chiều')
plt.colorbar(label='TenYearCHD')
plt.show()
```
